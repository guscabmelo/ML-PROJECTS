**ESTE SCRIPT FAZ A LEITURA DE UMA IMAGEM RBG E FAZA REDUÇÃO PARA GRAY SCAL E E PnB**

```python
# IMPORTANDO BIBLITECAS
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# DEFININDO AS FUNÇÕES PARA CALCULAR E CONVERTER A IMAGEM RGB PARA GRAY SCALE E BLACK 'N WHITE
def rgb_to_gray(rgb_array):
    # Converte uma imagem RGB para tons de cinza usando a fórmula ponderada
    gray_array = 0.299 * rgb_array[:, :, 0] + 0.587 * rgb_array[:, :, 1] + 0.114 * rgb_array[:, :, 2]
    return gray_array.astype(np.uint8)

def gray_to_bw(gray_array, threshold=128):
    # Converte uma imagem em tons de cinza para preto e branco usando um limiar
    bw_array = (gray_array > threshold) * 255
    return bw_array.astype(np.uint8)
```

```python
# CARREGANDO A IMAGEM
image_path = '/content/drive/MyDrive/FOTOS/DJI_0194.JPG'  # Substitua pelo caminho da sua imagem
image = Image.open(image_path)
rgb_array = np.array(image)
```

```python
# CONVERTENDO PARA GRAY SCALE
gray_array = rgb_to_gray(rgb_array)

# CONVERTENDO PARA BLACK 'N WHITE
bw_array = gray_to_bw(gray_array)
```

```python
# PLOTANDO A IMAGEM ORIGINAL E OS RESULTADOS
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# ORIGIBAL
axs[0].imshow(rgb_array)
axs[0].set_title('Imagem Original')
axs[0].axis('off')

# GRAY SCALE
axs[1].imshow(gray_array, cmap='gray')
axs[1].set_title('Imagem em Tons de Cinza')
axs[1].axis('off')

# BLACK 'N WHITE
axs[2].imshow(bw_array, cmap='gray')
axs[2].set_title('Imagem em Preto e Branco')
axs[2].axis('off')

plt.show()
```
